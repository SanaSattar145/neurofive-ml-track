```python
import seaborn as sns
import pandas as pd

# Seaborn se direct Titanic dataset load kar rahe hain (koi file ki zaroorat nahi)
df = sns.load_dataset('titanic')

# Missing values handle karna
df['age'] = df['age'].fillna(df['age'].median())
df.drop(columns=['deck'], inplace=True, errors='ignore')
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# Check karne ke liye
print(df.isnull().sum())
```

### Missing Values Handling Justification:
- **Age:** Missing values ko median se fill kiya hai taake extreme outliers ka asar na paray.
- **Embarked:** Jo choti moti missing values thin, unhein mode (most frequent value) se fill kar diya hai.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Style set kar rahe hain
sns.set_theme(style="whitegrid")

# 1. Histogram (Age distribution ke liye)
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], kde=True, color='blue')
plt.title('Histogram of Passenger Age')
plt.show()

# 2. Boxplot (Fare mein Outliers detect karne ke liye)
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['fare'], color='orange')
plt.title('Boxplot of Passenger Fare (Outlier Detection)')
plt.show()

# 3. Bar Chart (Survived count ke liye)
plt.figure(figsize=(6, 4))
sns.countplot(x='survived', data=df, palette='Set2')
plt.title('Bar Chart of Survival Count')
plt.show()

# 4. Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

### Question: Which feature do you think most affects survival, and why?
**Answer:**
The feature that most affects survival is **Sex** (Gender), because historical evacuation protocols prioritized women and children, resulting in significantly higher survival rates for females in the dataset. Additionally, **Pclass** (Passenger Class) strongly affected survival, as 1st class passengers had much better access to lifeboats and safety compared to 3rd class passengers.
